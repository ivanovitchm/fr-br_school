```python
# Load relevant libraries
import pandas as pd
import folium
from folium.plugins import HeatMap
from tqdm import tqdm
import numpy as np
from locale import atof
```

```python
# Import data
df_vehicle_0 = pd.read_csv("VehicleDataSet/vehicle_0.csv",sep=',')
df_vehicle_1 = pd.read_csv("VehicleDataSet/vehicle_1.csv",sep=',')
df_vehicle_2 = pd.read_csv("VehicleDataSet/vehicle_2.csv",sep=',')
df_vehicle_3 = pd.read_csv("VehicleDataSet/vehicle_3.csv",sep=',')
```

```python
df_vehicle_0.head()
```

```python
coordinates = []

for i in tqdm(range(len(df_vehicle_0))):
        coordinates.append([df_vehicle_0.ix[i,'LATITUDE'], 
                        df_vehicle_0.ix[i,'LONGITUDE'], 
                        df_vehicle_0.ix[i,'ENGINE_COOLANT_TEMP']])
    
    
#create map object
map_speed = folium.Map(
    location = [-5.791659, -35.228385],
    zoom_start= 12
)

max_speed = df_vehicle_0['ENGINE_COOLANT_TEMP'].max()
print(max_speed)

HeatMap(coordinates,max_val=max_speed).add_to(map_speed)
map_speed
```
